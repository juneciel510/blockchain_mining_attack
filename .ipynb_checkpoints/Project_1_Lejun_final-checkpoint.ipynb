### 2.2 Stubborn
#### 2.2.1 Theory  
The basic idea of stubborn attack is as follow:
<blockquote>Begin attack when detect there exist a fork with longest chain 1 block head comparing to its last published block, mining secretly, give up when the difference increase, release all secret block when its secret chain is 1 block ahead of the public chain, illustrated in the following figure. </blockquote>

### 2.3 A new implementation for GHOST protocol
#### 2.3.1 Theory
The basic idea of new implementation is as follow:

<blockquote>We set weight for each block. When a new block is generated, its initial weight is 1, the weight of all its parents will increase by 1. As a sequence, the weight of each block represents the block included in its sub tree.</br>

Thus, when we check for the heaviest sub tree, we simply by looking at weight of nodes. The idea of this idea is, when multiple blocks are added to the current heaviest subtree, the selfish miner need to publish the same number of its private block to make its own sub tree the heaviest one, thus win the competition. The o illustrated as following fig.</blockquote>

```python
#helper function for do statistics and plot
class statisticsNPlot:
    def __init__(self,system_parameter_dic,miners_parameter_dic,
                 testing_miningPower_dic=None,dostatistic=True      
                 ):
        print("---initial")
        self.system_parameter_dic=system_parameter_dic
        self.miners_parameter_dic=miners_parameter_dic
        self.testing_miningPower_dic=testing_miningPower_dic
        #temp store for statistics data for a certain difficulty
        self.df_statistics=pd.DataFrame()
        #to store all statistics data for different difficulty, key is difficulty
        self.df_statistics_dic={}
        self.attackers=[]
        self.miners=[]
        if dostatistic:
            self.doStatistics()

    def getAttackerNumber(self):
        return len(self.attackers)
       
    def getFieldStat(self,difficulty,name,field):
        return self.df_statistics_dic[difficulty].loc[[name],[field]]
    
     #plot statistics data for a certain difficulty for all wanted miners or attackers
    def plot_difficulty(self,difficulty,name_wantToShow,textstr=None):
        #print("---plot")
        plt.figure()
        x = np.linspace(0, 0.6, 100)
        y = np.linspace(0, 0.6, 100)
        color=["red","blue","green","yellow","purple"]
        plt.plot(x,y,color="black",label='x=y')
        i=0
        attacker_number=self.getAttackerNumber()         
        for name in name_wantToShow:
            x=self.df_statistics_dic[difficulty].loc[[name],['alpha']]
            y1=self.df_statistics_dic[difficulty].loc[[name],['fraction_in_chain']]
            y2=self.df_statistics_dic[difficulty].loc[[name],['fraction_lost']]
            line1=plt.plot(x, y1, color=color[i], marker='o',label='{} fraction_in_chain'.format(name))
            line2=plt.plot(x, y2, color=color[i],linestyle="dashed",marker='*',label='{} fraction_lost'.format(name))
            plt.title('{} v.s. alpha (difficulty={})'.format(self.system_parameter_dic["attack_type"],difficulty), fontsize=14)
            plt.xlabel('alpha/mining power ratio', fontsize=14)
            plt.ylabel('fraction', fontsize=14)
            i+=1

        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.grid(True)
        if  textstr!=None:   
            plt.text(0.65, 0.1, textstr, fontsize=10)
        plt.show()
    
        #plot statistics data of a miner or attacker for all difficulty
    def plot_name(self,name,textstr=None):
        plt.figure()
        x = np.linspace(0, 0.6, 100)
        y = np.linspace(0, 0.6, 100)
        color=["red","blue","green","yellow"]
        plt.plot(x,y,color="black",label='x=y')
        
        i=0
           
        for difficulty in self.system_parameter_dic["difficulty_list"]:
            x=self.df_statistics_dic[difficulty].loc[[name],['alpha']]
            y1=self.df_statistics_dic[difficulty].loc[[name],['fraction_in_chain']]
            y2=self.df_statistics_dic[difficulty].loc[[name],['fraction_lost']]
            line1=plt.plot(x, y1, color=color[i], marker='o',label='{} difficulty={} fraction_in_chain'.format(name,difficulty))
            line2=plt.plot(x, y2, color=color[i],linestyle="dashed",marker='*',label='{} difficulty={} fraction_lost'.format(name,difficulty))
            plt.title('same miner in different difficulty ({})'.format(self.system_parameter_dic["attack_type"]), fontsize=14)
            plt.xlabel('alpha/mining power ratio', fontsize=14)
            plt.ylabel('fraction', fontsize=14)
            i+=1


        plt.legend(loc='center left', bbox_to_anchor=(1, 0.7))
        plt.grid(True)
       
        if  textstr!=None:   
            plt.text(0.65, 0.1, textstr, fontsize=10)
        plt.show()
        
    #plot for all attackers
    def plotAllAttackers(self):
        for difficulty in self.df_statistics_dic:
            name_wantToShow=[item.name for item in self.attackers]
            self.plot_difficulty(difficulty,name_wantToShow)
            
    #display statistics data for all attackers
    def displayAllAttackersData(self): 
        self.displayParameters()
        for difficulty in self.df_statistics_dic:
            name_wantToShow=[item.name for item in self.attackers]
            self.displayStatistics(difficulty,name_wantToShow)
        
    #display pd dataframe statistics data for a certain difficulty and name list
    def displayStatistics(self,difficulty,name_list):
        display(self.df_statistics_dic[difficulty].loc[name_list,:])
            
    #display both system's and miners's parameters 
    def displayParameters(self): 
        display(pd.DataFrame([self.system_parameter_dic],index=["sys_parameters"]))
        df_miner_para = pd.DataFrame()
        copy=self.miners_parameter_dic.copy()

        if self.testing_miningPower_dic!=None:
            for attacker, testing_mining in self.testing_miningPower_dic.items():
                copy[attacker]["mining_power"]=testing_mining
        for name,para_dic in copy.items():
            df_miner_para=df_miner_para.append(pd.DataFrame([para_dic],index=[name]))
        display(df_miner_para)
        
    #show both data and plot for all attackers
    def showAllAttackers(self): 
        self.displayParameters()
        for difficulty in self.df_statistics_dic:
            name_wantToShow=[item.name for item in self.attackers]
            self.plot_difficulty(difficulty,name_wantToShow)
            self.displayStatistics(difficulty,name_wantToShow)
        
    #do statisctics for various difficulty at different mining ratio
    def doStatistics(self):
        print("---doing Statistics...")
        for difficulty in self.system_parameter_dic["difficulty_list"]:
            self.df_statistics=pd.DataFrame()
            testing_number=len(list(self.testing_miningPower_dic.values())[0])
            for i in range(testing_number):
                for name, miningPower_list in self.testing_miningPower_dic.items():
                    self.miners_parameter_dic[name]["mining_power"]=miningPower_list[i]
                self.calculateAverage(difficulty)

            self.df_statistics_dic[difficulty]=self.df_statistics
            
    
    def statForOneIteration(self,bc):   
      attack_type=self.system_parameter_dic["attack_type"]
      blockchain_size=self.system_parameter_dic["blockchain_size"]
    
      miners = []
      attacker=[]
      statistics={}
      networkPower={} 

      for name,parameter in miners_parameter_dic.items():
        for i in range(parameter["number"]):
            m = parameter["class"](parameter["mining_power"] ,name+str(i), bc)
            if parameter["class"].__name__=="Miner" and name!="big":
                miners.append(m)
            else:
                attacker.append(m)
            if parameter["network_power"]!=None:
                networkPower[m]=parameter["network_power"][i]
                
      self.attackers=attacker
      self.miners=miners

      if len(networkPower)>0:
        bc.setNetworkPowerWeight(networkPower)


      while bc.size < blockchain_size+1:
        if len(attacker)>0:
            for at in attacker:
              at.PoWSolver()

        for m in miners:
          m.PoWSolver()

        if len(attacker)>0:
            for at in attacker:
              at.UpdateLast()

        for m in miners:
          m.UpdateLast()

        if len(attacker)>0:
            #force publish all the private blocks
            for at in attacker:
                try:
                    for block in at.privateBlocks:
                        bc.add(block)
                    at.privateBlocks = []
                except:
                    pass                  

      total = 0
      minesNattacker=miners+attacker
        
      for ma in minesNattacker:
          total += ma.miningPower
     
      for item in minesNattacker:
        if bc.size-bc.lastBlock().height==0:
            fraction_lost=0
        else:
            fraction_lost=(bc.checkMinerAll(item)-bc.checkMiner(item)) /(bc.size-bc.lastBlock().height)
        statistics[item.name]={"blockchain_size":bc.size,
            "blockchain_height":bc.lastBlock().height,
                                   "fraction_in_chain":bc.checkMiner(item) /bc.lastBlock().height,
                                   "fraction_lost":fraction_lost,
                                   "alpha":item.miningPower/ total}
      return statistics
            
    #calculate averayge statistics of mulitple iterations for a certain difficulty and a certain attacker's mining power    
    def calculateAverage(self,difficulty):
        average_statistics = defaultdict(dict)
        iteration_times=self.system_parameter_dic["iteration_times"]
        blockchain_size=self.system_parameter_dic["blockchain_size"]

        for i in range(iteration_times):
          bc = Blockchain("0"+str(i) , difficulty,self.system_parameter_dic["protocol"])
          #statistics data for one interation
          statistics=self.statForOneIteration(bc)
          for name,dic in statistics.items():
            for metrics,value in dic.items():
                try:
                    average_statistics[name][metrics]=average_statistics[name][metrics]+value
                except:
                    average_statistics[name][metrics]=0
                    average_statistics[name][metrics]=average_statistics[name][metrics]+value

        for name,dic in statistics.items():
           for metrics,value in dic.items():
            average_statistics[name][metrics]=round(average_statistics[name][metrics]/iteration_times,3)


        for name,dic in average_statistics.items():
            average_statistics[name]["difficulty"] = difficulty

        for name,dic in average_statistics.items():
            self.df_statistics=self.df_statistics.append(pd.DataFrame(dic,index=[name]))        


def plotMultiStatistics(SP_name_dic,title,textstr=None,baseline=True):
    # SP_name_dic is a dictionary like {SP1:["selfish0","m0"],SP2:["selfish0","m3"]}
    #the first entity of this dictionary is considered as the baseline data, if not, set the baseline=False
    color=["red","blue","green","orange","deeppink","mediumorchid"]
    difficulty_list=list(SP_name_dic.keys())[0].system_parameter_dic["difficulty_list"]
    attack_type=list(SP_name_dic.keys())[0].system_parameter_dic["attack_type"]
    for difficulty in difficulty_list:
        plt.figure()
        x = np.linspace(0, 0.6, 100)
        y = np.linspace(0, 0.6, 100)
        plt.plot(x,y,label='x=y',color="black")
        if baseline:
            i=0
        else:
            i=1
        j=0
        for SP,name_list in SP_name_dic.items():
            for name in name_list:
                # attacker_name=SP.df_statistics_dic[difficulty].index.values[0]
                x=SP.df_statistics_dic[difficulty].loc[[name],['alpha']]
                y1=SP.df_statistics_dic[difficulty].loc[[name],['fraction_in_chain']]
                y2=SP.df_statistics_dic[difficulty].loc[[name],['fraction_lost']]
                # if it is the first data, it is the baseline data
                if i==0:
                    label1='{}, fraction_in_chain, Baseline'.format(name)
                    label2='{}, fraction_lost, Baseline'.format(name)
                else:
                    label1='{}, fraction_in_chain, Case_{}'.format(name,i)
                    label2='{}, fraction_lost, Case_{}'.format(name,i)
                line1=plt.plot(x, y1, color=color[j], marker='o',label=label1)
                line2=plt.plot(x, y2, color=color[j],linestyle="dashed",marker='*',label=label2)
                j+=1
            i+=1
           

        plt.title('{}  (difficulty={})'.format(title,difficulty), fontsize=14)
        plt.xlabel('alpha/mining power ratio', fontsize=14)
        plt.ylabel('fraction', fontsize=14)
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.50))
        plt.grid(True)
        if  textstr!=None:   
            plt.text(1.1, 0.10, textstr, fontsize=10)
        plt.show()
```

```python
#implementation for Block, Blockchain, Miner, SelfishMiner, StubbornMiner
# an new implementation for GHOST protocol, 
#which is faster than the previous recursive function when looking for the heaviest sub chain
class Block:
    def __init__(self, data, creator=None, previous=None, nonce=0):
        self.data = data
        if previous is None:
            self.previous = None
            self.previous_hash = ""
            self.creator = Miner(0 , "0")
            self.height = 0
            self.allParent = []
        else:
            self.previous = previous
            self.previous_hash = previous.hash
            self.creator = creator
            self.height = previous.height+1
            if self.previous.creator.name!="0":
             self.allParent =self.previous.allParent+ [previous]
            else:
                self.allParent =[previous]
        self.nonce = nonce
        self.hash = self.hash_block()
        self.children = []
        #a block's initial weight is 1, when a child is add, the weight increases by 1
        self.weight=1

        

    def hash_block(self):
        return hashbits(self.data+ self.creator.name + self.previous_hash + str(self.nonce))
    
    def updateWeight(self):
        for parent in self.allParent:
            parent.weight=parent.weight+1

    def print(self):
      print(self.data + " "+ self.creator.name + " " + str(self.height))

class Miner:
  def __init__(self, miningPower, name, blockchain=None):
    if miningPower==None:
        print("{}'s mining power is None".format(name))
    self.miningPower = miningPower
    self.nonce = random.randint(0,100000)
    self.name = name
    self.blockchain = blockchain
    if self.blockchain != None:
      self.lastBlock = blockchain.lastBlock()
  
  def UpdateLast(self):
    latest = self.blockchain.lastBlock()
    if latest.height > self.lastBlock.height:
        self.lastBlock = latest

  def PoWSolver(self):
    for i in range (0 , self.miningPower):
      newBlock = Block(str(self.blockchain.size), self, self.lastBlock, self.nonce)
      h = newBlock.hash_block()
      count = numberOfInitZeros(h)
      if count >= self.blockchain.difficulty:
        self.blockchain.add(newBlock)
        self.lastBlock = newBlock
      self.nonce += 1

#blockchain provide three protocol:
#1."longest_noNetworkPower": follow longest chain rule, do not model any power network, 
#2."longest_noNetworkPower": follow longest chain rule, miners and attacker can be assigned with different power network,
#3. "ghost_new" :follow new implemented GHOST protocol
#4. "ghost_original" :follow GHOST protocol from teaching material
class Blockchain:
    def __init__(self, genesis_data, difficulty, protocol="longest_noNetworkPower"):
        self.chain = []
        self.chain.append(Block(genesis_data))
        self.difficulty = difficulty
        self.size = 0
        self.protocol = protocol
        #a dictionary,miner as key, network power as value
        self.networkPower={}
    
    def setNetworkPowerWeight(self,networkPower):
        self.networkPower=networkPower
        #print("self.networkPower",self.networkPower)

    def lastBlock(self):
        if self.protocol == "longest_noNetworkPower":
            return self.longestChain_noNetworkPower()
        if self.protocol == "longest_withNetworkPower":
            return self.longestChain_withNetworkPower()
        if self.protocol == "ghost_new":
            return self.heaviestChain_new(self.chain[0])
        if self.protocol == "ghost_original":
            return self.heaviestChain_orginal(self.chain[0])
        print("Invalid protocol")
        return None
    
    # no network power modelled
    def longestChain_noNetworkPower(self): 
      #print("---no network power modelled")
      max = self.chain[0]
      for block in self.chain:
        if block.height > max.height:
          max = block
      return max
    
    #model net network power
    def longestChain_withNetworkPower(self): 
      max = self.chain[0].height
      for block in self.chain:
        if block.height > max:
          max = block.height
      maxes = [block for block in self.chain if block.height == max]
      #if it is during the initialization or when the blockchain just have the genesis block
      if len(maxes)==1 or len(self.networkPower)==0 or block.creator.name=='0':
            r = random.choices(maxes, k=1)
      else:
          networkPower_weight=[]
          for block in maxes:
            networkPower_weight.append(self.networkPower[block.creator])   
          r = random.choices(maxes,weights =networkPower_weight, k=1)
        
      return r[0]


    def heaviestChain_new(self, root):
      if len(root.children) == 0:
        return root
      else:
        max = root.children[0]
        maxChilds = max.weight
        for block in root.children:
          tmp = block.weight
          if tmp > maxChilds:
            max = block
            maxChilds = tmp
        return self.heaviestChain_new(max)

    def heaviestChain_orginal(self, root):
      if len(root.children) == 0:
        return root
      else:
        max = root.children[0]
        maxChilds = self.numberOfBlocksInSubtree(max)
        for block in root.children:
          tmp = self.numberOfBlocksInSubtree(block)
          if tmp > maxChilds:
            max = block
            maxChilds = tmp
        return self.heaviestChain_orginal(max)

    def numberOfBlocksInSubtree(self, block):
      if len(block.children) == 0:
        return 1
      else:
        sum = 0
        for b in block.children:
          sum += self.numberOfBlocksInSubtree(b)
        return 1 + sum

    def add(self, newBlock):
        self.chain.append(newBlock)
        newBlock.previous.children.append(newBlock)
        self.size +=1
        #when a block is added, all its parents should increase their weight by 1
        newBlock.updateWeight()
    
    
    def checkMiner(self, miner):
      last = self.lastBlock()
      count = 0
      while last!=None:
        if last.creator == miner:
          count += 1
        last = last.previous
      return count

    def hasFork(self):
      for block1 in self.chain:
        for block2 in self.chain:
          if block1!=block2 and block1.height == block2.height:
            return True
      return False

    def checkMinerAll(self, miner):
      count = 0
      for block in self.chain:
        if block.creator == miner:
          count += 1
      return count
        
    def print(self):
      for block in self.chain:
        block.print()
        print("________")

class SelfishMiner(Miner):
  def __init__(self, miningPower, name, blockchain=None):
    super().__init__(miningPower, name, blockchain)
    self.privateBlocks = []
    self.publishNext = False

  def UpdateLast(self):
    latest = self.blockchain.lastBlock()
    publicheight = latest.height
    if publicheight > self.lastBlock.height:
        self.privateBlocks = []
        self.lastBlock = latest
        self.publishNext = False
    if publicheight == self.lastBlock.height-1 and len(self.privateBlocks)> 1:
        for block in self.privateBlocks:
            self.blockchain.add(block)
        self.privateBlocks = []
    if publicheight == self.lastBlock.height:
        for block in self.privateBlocks:
            self.blockchain.add(block)
        self.privateBlocks = []
        self.publishNext = True

  def PoWSolver(self):
    for i in range (0 , self.miningPower):
      newBlock = Block(str(self.blockchain.size), self, self.lastBlock, self.nonce)
      h = newBlock.hash_block()
      count = numberOfInitZeros(h)
      if count >= self.blockchain.difficulty:
        if self.publishNext:
            self.blockchain.add(newBlock)
            self.publishNext = False
        else:
            self.privateBlocks.append(newBlock)
        self.lastBlock = newBlock
      self.nonce += 1
    
class StubbornMiner(Miner):
  def __init__(self, miningPower, name, blockchain=None):
    super().__init__(miningPower, name, blockchain)
    self.privateBlocks = []
    self.publishNext = True

#   def UpdateLast(self):
#     latest = self.blockchain.lastBlock()
#     publicheight = latest.height
#     if publicheight-1 > self.lastBlock.height:
#       #give up attack
#         self.privateBlocks = []
#         self.lastBlock = latest
#         self.publishNext = True
#     if publicheight-1 == self.lastBlock.height:
#       #begin attack and develope on its own secret chain
#         self.publishNext = False
#     if publicheight+1 == self.lastBlock.height :
#       #attack win, publish block
#         for block in self.privateBlocks:
#             self.blockchain.add(block)
#         self.privateBlocks = []
#         self.publishNext = True
        
  def UpdateLast(self):
    latest = self.blockchain.lastBlock()
    publicheight = latest.height
    #attacker lose
    if publicheight-1 > self.lastBlock.height:
        self.privateBlocks = []
        self.lastBlock = latest
        self.publishNext = True
    
    #initial state
    fork_height=publicheight-1 
    fork_blocks = [block for block in self.blockchain.chain if block.height == fork_height]
    attacker_block=[block for block in fork_blocks if block.creator.name ==self.name]
    if len(fork_blocks)>1 and len(attacker_block)==1:
        #begin attack and develope on its own secret chain
        self.publishNext = False
        
    #still keep secret block
    if publicheight == self.lastBlock.height :
        self.publishNext = False
    
    #attack win, publish block
    if publicheight+1 <= self.lastBlock.height :
        for block in self.privateBlocks:
            self.blockchain.add(block)
        self.privateBlocks = []
        self.publishNext = False

  def PoWSolver(self):
    for i in range (0 , self.miningPower):
      newBlock = Block(str(self.blockchain.size), self, self.lastBlock, self.nonce)
      h = newBlock.hash_block()
      count = numberOfInitZeros(h)
      if count >= self.blockchain.difficulty:
        if self.publishNext:
            self.blockchain.add(newBlock)
            #self.publishNext = False
        else:
            self.privateBlocks.append(newBlock)
        self.lastBlock = newBlock
      self.nonce += 1
```

```python
##adapt the show function to show the weight for every block
####### Drawing blockchain, not important
def maxHeight(parent):
  if len(parent.children) == 0:
    return parent.height 
  max = 0
  for child in parent.children:
    m = maxHeight(child)
    if m> max:
      max = m
  return max


def drawBlockchain(parent, level, html, parentLevel, childN = 0, total = 0):
  color = "#AEF751"
  if parentLevel!=-1:
    color = "#7EDBF6"
    if type(parent.creator) is SelfishMiner:
      color = "#F58A70"
    #elif type(parent.creator) is SelfishMiner:
    #  color = "#F59AEE"
  parent.children.sort(key=lambda x: (maxHeight(x)), reverse=True)
  xx = childN
  level += childN
  html += '<g>'
  html += '<rect x="'+str(30+ 100*parent.height)+'" y="'+str(30+ 100*level)+'" width="60" height="60" stroke="black" stroke-width="1" fill="'+color+'" />'
  html += '<text x="'+str((60+ 100*parent.height))+'" y="'+str((60+ 100*level))+'" dominant-baseline="middle" text-anchor="middle" font-family="Verdana" font-size="10" font-weight="bold" fill="black">'+str(parent.creator.name)+":"+str(+parent.weight)+'</text>'
  if parentLevel != -1:
    if (parent.previous.children.index(parent)) == 0:
      html += '<line stroke-width="1px" stroke="#000000"  x1='+str(30+ 100*parent.height)+' y1="'+str(60+ 100*level)+'" x2="'+str(95+ 100*parent.previous.height)+'" y2="'+str(60+ 100*parentLevel)+'" style="marker-end: url(#markerArrow)"/>'
    else:
      html += '<line stroke-width="1px" stroke="#000000"  x1='+str(30+ 100*parent.height)+' y1="'+str(60+ 100*level)+'" x2="'+str(65+ 100*parent.previous.height)+'" y2="'+str(95+ 100*parentLevel)+'" style="marker-end: url(#markerArrow)"/>'
  html += '</g>'
  l = level
  childN = 0
  for child in parent.children:
    html,n, t = drawBlockchain(child, l, html, level, childN, total)
    if n > 0:
      childN += n
    if t > 0:
      total += t
    l = l+1
  return html, childN+ len(parent.children)-1, total+ len(parent.children)-1


def show(bc):
  htmll = ""
  html = ""
  htmll, n, t = drawBlockchain(bc.chain[0], 0, html, -1)
  html = '<svg height="'+str(115*(n+1))+'" width="'+str(115*maxHeight(bc.chain[0]))+'">'
  html += '<defs><marker id="markerArrow" markerWidth="10" markerHeight="10" refX="2" refY="6" orient="auto"><path d="M2,2 L2,11 L10,6 L2,2" style="fill: #000000;" /> </marker> </defs>'
  html += htmll
  html += '</svg>'
  display(HTML(html))
```

###  10.3 Senario 2: attackers-->  longest chain rule, miners --> GHOST_new

#### 10.3.1 Implementation 

```python
#selfish follow the longest chain rule
class SelfishMiner(Miner):
  def __init__(self, miningPower, name, blockchain=None):
    super().__init__(miningPower, name, blockchain)
    self.privateBlocks = []
    self.publishNext = False
    if self.blockchain != None:
        self.lastBlock = blockchain.longestChain_noNetworkPower()
    
  def UpdateLast(self):
    latest = self.blockchain.longestChain_noNetworkPower()
    publicheight = latest.height
    if publicheight > self.lastBlock.height:
        self.privateBlocks = []
        self.lastBlock = latest
        self.publishNext = False
    if publicheight == self.lastBlock.height-1 and len(self.privateBlocks)> 1:
        for block in self.privateBlocks:
            self.blockchain.add(block)
        self.privateBlocks = []
    if publicheight == self.lastBlock.height:
        for block in self.privateBlocks:
            self.blockchain.add(block)
        self.privateBlocks = []
        self.publishNext = True

  def PoWSolver(self):
    for i in range (0 , self.miningPower):
      newBlock = Block(str(self.blockchain.size), self, self.lastBlock, self.nonce)
      h = newBlock.hash_block()
      count = numberOfInitZeros(h)
      if count >= self.blockchain.difficulty:
        if self.publishNext:
            self.blockchain.add(newBlock)
            self.publishNext = False
        else:
            self.privateBlocks.append(newBlock)
        self.lastBlock = newBlock
      self.nonce += 1
```

#### 10.4.1 Implementation

```python
#a better approach??
class SelfishMiner(Miner):
  def __init__(self, miningPower, name, blockchain=None):
    super().__init__(miningPower, name, blockchain)
    self.privateBlocks = []
    self.publishNext = False

  def GetBlockSameHeight(self,privateBlock):
        sameHeightNodes=[block for block in self.blockchain.chain if block.height==privateBlock.height]
        return sameHeightNodes    
    
  def UpdateLast(self):
    latest = self.blockchain.lastBlock()
    publicheight = latest.height
    
    if len(self.privateBlocks)==0:
        self.lastBlock = latest
    else:
        sameHeightBlocks=self.GetBlockSameHeight(self.privateBlocks[0])
        if len(sameHeightBlocks)!=0:
            lead=[self.privateBlocks[0].weight-block.weight for block in sameHeightBlocks]
            minLead=min(lead)
            if minLead==0:  
                for block in self.privateBlocks:
                    self.blockchain.add(block)
                    self.privateBlocks = []
                    self.publishNext = True
                    self.lastBlock = latest
            if minLead==1 or minLead==2: 
                self.publishNext = False
                for i in range(2):
                    self.blockchain.add(self.privateBlocks.pop(0))
            if minLead>=3: 
                sameHeightBlocks=self.GetBlockSameHeight(self.privateBlocks[0])
                weight=[block.weight for block in sameHeightBlocks]
                totalWeight=sum(weight)
                for i in range(totalWeight):
                    if len(self.privateBlocks)>0:
                        self.blockchain.add(self.privateBlocks.pop(0))

  def PoWSolver(self):
    for i in range (0 , self.miningPower):
      newBlock = Block(str(self.blockchain.size), self, self.lastBlock, self.nonce)
      h = newBlock.hash_block()
      count = numberOfInitZeros(h)
      if count >= self.blockchain.difficulty:
        if self.publishNext:
            self.blockchain.add(newBlock)
            self.publishNext = False
        else:
            self.privateBlocks.append(newBlock)
        self.lastBlock = newBlock
      self.nonce += 1    
```
